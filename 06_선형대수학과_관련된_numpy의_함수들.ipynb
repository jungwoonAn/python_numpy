### 1. 행렬 다루기
- 행렬곱: **@**또는 **dot()**  
- 전치행렬: A.**T**  
- 역행렬(Inverse of a matrix) : **np.linalg.inv(A)**  
- 단위행렬(Identity matrix) : **np.eye(n)**  
- 대각행렬(Diagonal matrix) : **np.diag(A)**  
- 대각합(Trace) : **np.trace(A)**  
- 행렬식(Matrix Determinant) : **np.linalg.det(A)**  
- 연립방정식해풀기(Solve a linear matrix equation) : **np.linalg.solve(A, B)**  
- 고유값(Eigenvalue), 고유벡터 (Eigenvector) : **w, v = np.linalg.eig(A)**  
- 특잇값분해(Singular Value Decomposition) : **u, s, vT = np.linalg.svd(A)**  

```python
import numpy as np
```

```python
A = np.array([[1,2], [3,4]])
```

```python
A
```

##### 8) 연립방정식 해풀기 : solve()는 선형방정식을 구함
    Ax = B 라는 행렬식에서 x 행렬을 알기 위해서는 양쪾에 A의 역행렬을 곱해야 함

```python
x = [2,3]
y = [6.8, 7.3]
```

```python
A = np.c_[x, np.ones(2)]
```

```python
A
```

```python
B = np.array(y)
```

```python
B
```

```python
np.linalg.solve(A, B)
```

=> y = wx + b (학교에서 y=ax+b)  
위 결과에서 w = 0.5(weight, 가중치=기울기), b = 5.8(bias, 편향=y절편)

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, np.multiply(x, 0.5) + 5.8)
plt.show()
```

#### 12) 선형 연립 방정식
선형연립방정식.png

```python
X = [32,64,96,118,126,144,152,158]
Y = [17,24,62,49,52,105,130,125]
```

```python
A = np.c_[X, np.ones(len(X))]
```

```python
A
```

```python
B = np.array(Y)
```

```python
B
```

```python
# 의사역행렬
inv_A = np.linalg.inv(A.T @ A) @ A.T
```

```python
w, b = inv_A @ B
```

```python
w, b
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X, Y)
plt.show()
```

```python
plt.scatter(X, Y)
plt.plot(X, np.multiply(X, w)+b)
plt.show()
```
