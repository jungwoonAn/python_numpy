### 1. 넘파이
- 넘파이 : 넘파이(Numpy)는 파이썬을 사용한 과학 컴퓨팅의 기본 패키지  
- 동종 모음 : 넘파이의 주요 객체는 동종(homogeneous)의 다차워 배열(요소들은 모두 같은 타입)  
- 넘파이 기능 : N차원 배열 객체 생성 및 관리  
  선형 대수학(Linear algebra), 푸리에 변환(Fourier transform) 기능  
  난수(Random number)생성 기능  
- 축 : 넘파이의 차원들은 축(axis)으로 불림
- 공식 사이트 : https://www.numpy.org/

### 3. ndarray 속성  
- **ndarray.ndim** : 배열의 **축(Axis) 수**, 차원  
- **ndarray.shape** : 각 차원의 **배열 크기**를 나타내는 정수 타입의 튜플  
  shape는 (n,m)형태. 행렬은 n개의 행과  m개의 열.  
  shape 튜플의 길이는 축의 수(ndim)
- **ndarray.size** : 배열의 **요소의 총 수**, shape의 각 요소의 곱과 동일  
- **ndarray.dtype** : 배열 내의 **요소의 타입**  
  형변환의 개념이 아님. 지종한 타입의 크기만큼 잘라서 해당 타입으로 인식함.  
  형변환은 astype(t)함수를 이용
- **ndarray.itemsize** : 배열의 각 요소의 바이트 단위 사이즈(요소가 저장하는 데이터의 크기)

```python
import numpy as np
```

```python
A = np.arange(15).reshape(3, 5)  # arange() : 배열생성 0~14, reshape() : 3행 5열 2차원
```

```python
A
```

```python
A.ndim  # 배열의 축(Axis) 수, 차원
```

```python
A.shape  # 축의 크기
```

```python
A.size  # 요소의 수
```

```python
A.dtype  # 요소의 타입
```

```python
A.itemsize  # 바이트 단위의 사이즈 => 요소가 저장하는 데이터의 크기
```

### 3. dtype의 이해
- dtype 속성으로 타입을 지정하면 기존의 데이터들이 새로 바뀌는 타입의 크기만큼 읽혀짐
- 처음의 배열은 shape가(3, 4)이고 아이템들의 타입이 int32였다면 이후 dtype을 int64로 지정하면 정수 두 개값이 하나의 정수로 읽혀짐

```python
A = np.arange(12).reshape(3,4)
A
```

```python
A.dtype
```

```python
A.dtype = np.int32
A
```
